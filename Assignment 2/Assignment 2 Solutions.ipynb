```python
x = np.linspace(0,10, 1000)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), linestyle = "dashed")
```

```python
x = np.linspace(0.1,10,1000)
plt.yscale("log")
plt.plot(x, np.log(x))
plt.plot(x, np.exp(x))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,5))
x = np.linspace(0.1, 10, 1000)

ax1.plot(x, 0 * x)
ax1.plot(x, np.exp(x))

ax2.plot(x, np.log(x))
plt.yscale("log")
ax2.plot(x, np.exp(x))
```
